## 波士頓房價預測

```python
%matplotlib inline
import numpy as np
from sklearn import datasets
from sklearn.linear_model  import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

np.random.seed(29)

#讀取波士頓房價
boston= datasets.load_boston()
print("資料及的特徵欄位名稱：",boston.feature_names)
x=boston.data
y=boston.target


#80%當作訓練集，20%當作測試集
x_train, x_test, y_train, y_test=train_test_split(x,y ,test_size=0.2)
#線性迴歸
regression_clf=LinearRegression()
regression_clf.fit(x_train,y_train)
y_predict=regression_clf.predict(x_test)
score=r2_score(y_test, y_predict)
print("房價預測準確率：",score*100)

#真實房價與預測房價對比圖
plt.plot(y_test,label='Real Price')
plt.plot(y_predict,label='Predic Price')
plt.legend()
```

## 鳶尾花分類

```python
import numpy as np
from sklearn import datasets
from sklearn.tree  import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

np.random.seed(47)


#讀取Iris分類
iris= datasets.load_iris()
print("資料及的特徵欄位名稱：",iris.feature_names)
x=iris.data
y=iris.target


#2/3當作訓練集，1/3當作測試集
x_train, x_test, y_train, y_test=train_test_split(x,y ,test_size=0.33)
#決策樹
decision_tree_clf=DecisionTreeClassifier(criterion='entropy')
decision_tree_clf.fit(x_train,y_train)
y_predict=decision_tree_clf.predict(x_test)
score=accuracy_score(y_test, y_predict)
print("鳶尾花分類預測準確率：",score*100)
```

```python
from sklearn.tree import export_graphviz
import graphviz
feature_names=['sepal length','sepal width','petal length','petal width']

#將模型匯出可繪圖格式
dot_data= export_graphviz(decision_tree_clf, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True)  

#繪圖
from sklearn.tree import export_text
from sklearn.tree import DecisionTreeClassifier, plot_tree




plt.figure(figsize=(16, 16))
plot_tree(decision_tree_clf, feature_names=feature_names,  
                         class_names=iris.target_names, filled=True, rounded=True, proportion=True)
plt.savefig('tree.jpg') 
```
